# Analysis 

# Number of medicine available of different benefit area

```python
medicine_counts = x2.groupby('key_benefit_areas')['name'].count()
print(medicine_counts)
```

# Price range of medicine for each benefit area.  

```python
df = pd.merge(x1,x2, how = 'inner', on = 'name')
```

```python
price_range = df.groupby('key_benefit_areas')['selling_price'].agg(['min', 'max'])

print(price_range)
```

```python
plt.figure(figsize = (12,7))
df['key_benefit_areas'].value_counts().plot(kind = 'line')
plt.title("MEDICINE COUNTS")
plt.ylabel("key_benefit_areas")
plt.legend(["key_benefit_areas"])
plt.show()
```

# Brand specialization(Key Benefits) of each area. 

```python
brand_specialization = x2.groupby('key_benefit_areas')['brand_names'].unique()
print(brand_specialization)
print(len(brand_specialization))
```

# Average price, min price , max price and number of products for each brand.

```python
df = pd.merge(x1,x2, how = 'inner', on = 'name')
brand_metrics = df.groupby('brand_names')['selling_price'].agg(['mean', 'min', 'max', 'count'])
brand_metrics.columns = ['Average Price', 'Min Price', 'Max Price', 'Number of Products']
print(brand_metrics)
```

# Average number of rating for each brand in their specialization products where  the number of rating is not NULL.

```python
df[df['number_of_ratings'] > 0].groupby('brand_names').agg({'number_of_ratings':'mean'})
```

```python
plt.figure(figsize = (15,7))
df['brand_names'].value_counts().plot(kind = 'bar')
plt.title("Brand specification")
plt.ylabel("number_of_ratings")
plt.legend(["number_of_ratings"])
plt.show()
```

# Maximum number of times ingredient used in each benefit area.

```python
a = x2.groupby('key_benefit_areas').agg({"key_ingredients":"max"})
a
```

```python
plt.figure(figsize = (12,7))
x2['key_benefit_areas'].value_counts().plot(kind = 'bar')
plt.title("ingredients")
plt.xlabel("benifits")
plt.ylabel("key_ingredients")
plt.legend(["key_ingredients"])
plt.show()
```

# Average cost for ingredients

```python
df = pd.merge(x1,x2, how = 'inner', on = 'name')
price_range = df.groupby('key_ingredients')['selling_price'].agg(['mean']).astype("int64")
print(price_range)
```

```python
plt.figure(figsize = (12,7))
x2['key_ingredients'].value_counts().plot(kind = 'hist')
plt.title("ingredients")
plt.xlabel("ing.")
plt.ylabel("selling_price")
plt.legend(["selling_price"])
plt.show()
```

```python
df['selling_price'].plot(kind='box', title='price')
plt.show()
```

# SELF ANALYSIS

What are the top 10 most commonly prescribed medications?

```python
df.groupby('name').agg({'number_of_ratings':'count'}).sort_values(by='number_of_ratings', ascending = False).iloc[:10]
```

Which medications have the highest and lowest prices?

```python
df = pd.merge(x1,x2, how = 'inner', on = 'name')
```

```python
lowest_price_medication = df.iloc[0]['name']
lowest_price = df.iloc[0]['selling_price']
```

```python
highest_price_medication = df.iloc[-1]['name']
highest_price = df.iloc[-1]['selling_price']
```

```python
print(lowest_price_medication ," = ", lowest_price)
print(highest_price_medication ," = ", highest_price)
```

What are the top 10 medication brands or manufacturers with the highest number of available products?

```python
x2.groupby("brand_names").agg({"name":"count"}).sort_values('name',ascending=False).iloc[:10]
```

Are there any correlations between medication ratings and their prices?

```python
correlation = df['ratings'].corr(df['selling_price'])
print(correlation)
```

```python
df.plot(x='ratings',y='selling_price',kind='scatter')
```

Which Medicine is least popular among the all and what is the possible reason for that?
